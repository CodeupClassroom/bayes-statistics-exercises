# Binomial vs Poisson

> A marketing website has an average click-through rate of 2%. One day they observe 4326 visitors and 97 click-throughs. How likely is it that this many people or more click through?


```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 4326
p = .02

x = np.arange(0, 200)
y_binom = stats.binom(n, p).pmf(x)
y_poisson = stats.poisson(n * p).pmf(x)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y_binom)
plt.title('Binomial Model')
plt.subplot(2, 1, 2)
plt.plot(x, y_poisson)
plt.title('Poisson Model')
```

```python
plt.plot(x, y_binom, label='binom')
plt.plot(x, y_poisson, label='poisson')
plt.legend()
```

## A more illustrative example

```python
import seaborn as sns
import pandas as pd
```

```python
n = 3
p = .5

x = np.arange(0, 7)
y_binom = stats.binom(n, p).pmf(x)
y_poisson = stats.poisson(n * p).pmf(x)

pd.DataFrame({
    'Binomial Model\nn = 3, p = .5': y_binom,
    'Poisson Model\n$\lambda$ = 3 x .5 = 1.5': y_poisson
}, index=x).plot.bar(width=.9)
plt.title('Comparison of Binomial vs Poisson Model')
plt.xlabel('x')
plt.ylabel('P(X = x)')
```

## Conclusion

- for a large n, there's not much difference between the two
- for a small n, there is a big difference
- choose a poisson when there's no upper limit on number of events or number of successes
- choose binomial when there is an upper bound on number of successes
